```python
import pandas as pd
```

```python
df=pd.read_csv('Advertising.csv',index_col=0)
df.head()
```

```python
df.tail()
```

```python
df.shape
```

```python
import seaborn as sns
%matplotlib inline
```

```python
sns.pairplot(df,x_vars=['TV','Radio','Newspaper'], y_vars='Sales',kind='reg',size=7, aspect=0.7)
```

```python
feature_col=['TV','Radio']
X=df[feature_col]
X.head()
```

```python
print(X.shape)
print(type(X))
```

```python
y=df['Sales']
y.head()
```

```python
from sklearn.cross_validation import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
print(model.intercept_)
print(model.coef_)
```

```python
list(zip(feature_col,model.coef_))
```

```python
y_pred=model.predict(X_test)
```

```python
import numpy as np
from sklearn import metrics
print(np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
```
